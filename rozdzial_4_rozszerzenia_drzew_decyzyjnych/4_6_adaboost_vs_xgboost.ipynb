# 4.6 AdaBoost vs XGBoost

```python
import numpy as np
```

```python
total_error = np.linspace(0, 1, 100)
total_error
```

```python
eps = 10e-5
```

```python
amount_of_say = 0.5 * np.log((1 - total_error + eps) /
                             (total_error + eps))
amount_of_say
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(total_error, amount_of_say)
```

```python
_, ax = plt.subplots(nrows=1, ncols=3, figsize=(10, 5))
for i, w_init in enumerate([0.001, 0.1, 0.5]):
    ax[i].plot(amount_of_say, w_init * (np.e ** amount_of_say))
```

```python
_, ax = plt.subplots(nrows=1, ncols=3, figsize=(12, 5))
for i, w_init in enumerate([0.001, 0.1, 0.5]):
    ax[i].plot(amount_of_say, w_init * (np.e ** (-amount_of_say)
                                        ))
```
